# 1. Métricas clave

```python
# Función para calcular porcentaje de tiros anotados
def calcular_porcentaje_tiros(anotados, intentados):
    return (anotados / intentados) * 100 if intentados != 0 else 0

# Tiros de campo (T2 + T3)
tiros_totales_intentados = len(df_main)
tiros_totales_anotados = len(df_main[df_main['SHOT_MADE_FLAG'] == 1])
porcentaje_tiros_campo = calcular_porcentaje_tiros(tiros_totales_anotados, tiros_totales_intentados)
print(f"%TC: {porcentaje_tiros_campo:.2f}")

# Tiros de 2 y tiros de 3
df_tiros_dos = df_main[df_main['SHOT_TYPE'] == '2PT Field Goal']
df_tiros_tres = df_main[df_main['SHOT_TYPE'] == '3PT Field Goal']

tiros_dos_intentados = len(df_tiros_dos)
tiros_tres_intentados = len(df_tiros_tres)
tiros_dos_anotados = len(df_tiros_dos[df_tiros_dos['SHOT_MADE_FLAG'] == 1])
tiros_tres_anotados = len(df_tiros_tres[df_tiros_tres['SHOT_MADE_FLAG'] == 1])

porcentaje_tiros_dos = calcular_porcentaje_tiros(tiros_dos_anotados, tiros_dos_intentados)
porcentaje_tiros_tres = calcular_porcentaje_tiros(tiros_tres_anotados, tiros_tres_intentados)

print(f"%T2: {porcentaje_tiros_dos:.2f}")
print(f"%T3: {porcentaje_tiros_tres:.2f}")

# Puntos totales y ratio de tiros libres
puntos_totales = 38389
puntos_TC = 2 * tiros_dos_anotados + 3 * tiros_tres_anotados
puntos_TL = puntos_totales - puntos_TC
ratio_TL = (puntos_TL / puntos_totales) * 100

print(f"Ratio TL: {ratio_TL:.2f}")
```

```python
# Métricas avanzadas

# Effective Field Goal Percentage
EFG = (int(tiros_dos_anotados) + 1.5*int(tiros_tres_anotados))/int(tiros_totales_intentados)
print("%eFG: ", 100*EFG)

# 2 Point Rate
twoPR = (int(tiros_dos_intentados))/int(tiros_totales_intentados)
print("2PR: ", twoPR*100)

# 3 Point Rate
threePR = (int(tiros_tres_intentados))/int(tiros_totales_intentados)
print("3PR: ", threePR*100)

# Points Per Attempt
puntos = 2*int(tiros_dos_anotados) + 3*int(tiros_tres_anotados)
PPA = puntos/int(tiros_totales_intentados)
print("PPA: ", PPA)
```

# 2. Distribuciones

```python
# Función para calcular porcentajes de tiro (intentados y anotados) según la zona
def calcular_porcentaje_tiros_por_zona(df, zona, total_tiros):
    tiros_zona = df[df['SHOT_ZONE_BASIC'] == zona]
    porcentaje_intentados = (len(tiros_zona) / total_tiros) * 100
    tiros_anotados = len(tiros_zona[tiros_zona['SHOT_MADE_FLAG'] == 1])
    porcentaje_anotados = (tiros_anotados / len(tiros_zona) * 100) if len(tiros_zona) > 0 else 0
    return porcentaje_intentados, porcentaje_anotados

# Total de tiros intentados
tiros_totales_intentados = len(df_main)

# Definir zonas de tiro
zonas_tiro = ['Mid-Range', 'In The Paint (Non-RA)', 'Restricted Area', 'Backcourt', 
              'Above the Break 3', 'Left Corner 3', 'Right Corner 3']

# Calcular y mostrar porcentajes de tiros por zona
for zona in zonas_tiro:
    porcentaje_intentados, porcentaje_anotados = calcular_porcentaje_tiros_por_zona(df_main, zona, tiros_totales_intentados)
    print(f"{zona}: {porcentaje_intentados:.2f}% intentados, {porcentaje_anotados:.2f}% anotados")
```
